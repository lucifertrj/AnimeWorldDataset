```python
import pandas as pd
```

```python
one_piece = pd.read_html('https://listfist.com/list-of-one-piece-arcs')
```

```python
naruto = pd.read_html("https://listfist.com/list-of-naruto-arcs")
```

```python
bleach = pd.read_html("https://listfist.com/list-of-bleach-arcs")
```

```python
df1 = pd.DataFrame(one_piece[0])
df1['BigThree'] = 'One Piece'
df2 = pd.DataFrame(naruto[0])
df2['BigThree'] = 'Naruto'
df3 = pd.DataFrame(bleach[0])
df3['BigThree'] = 'Bleach'
```

```python
df1
```

```python
df2.rename(columns={"TotalPages(avg 19)":"TotalPages"},inplace=True)
```

```python
df3.rename(columns={"TotalPages(avg 19)":"TotalPages"},inplace=True)
```

```python
df = pd.concat([df1,df2,df3])
```

```python
df.sample(10)
```

```python
df.drop("#",inplace=True,axis=1)
```

```python
df.to_csv("BigThree.csv",index = False)
```
